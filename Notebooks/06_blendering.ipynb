```python
import pandas as pd
from joblib import load
```

```python
# Load the test set
test_df = pd.read_csv('../data/test.csv')

# Preprocess with same logic as during training
import sys
sys.path.append('../Scripts')
from processing import preprocess

test_clean = preprocess(test_df, add_log_price=False)
```

```python
# Load saved models
rf = load('../Results/random_forest_model.joblib')
xgb = load('../Results/xgboost_model.joblib')
```

```python
# Make predictions
rf_pred = rf.predict(test_clean)
xgb_pred = xgb.predict(test_clean)

# Blend predictions (simple average)
blend_pred = (rf_pred + xgb_pred) / 2
```

```python
# Create submission DataFrame
submission = pd.DataFrame({
    'ID': test_df['ID'],
    'price': blend_pred
})

# Save to Results folder
submission.to_csv('../Results/submission_blend.csv', index=False)
```
