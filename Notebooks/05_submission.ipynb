Import libraries

```python
import pandas as pd
from joblib import load
```

Load and Preprocess the test data using the same function that we used for training. This ensures the features are consistent with those on whic the model was trained on.

```python
test_df = pd.read_csv('../data/test.csv')

# Preprocess using your existing logic
import sys
sys.path.append('../Scripts')
from processing import preprocess

# Match Random Forest setup â€” no log_price
test_clean = preprocess(test_df, add_log_price=False)
```

Load the Random Forest model saved in the Results folder

```python
model = load('../Results/random_forest_model.joblib')
```

Predict the car prices on the preprocessed test dataset

```python
predictions = model.predict(test_clean)
```

We create a csv file and export the results of the model to it for the Kaggle submission.

```python
submission = pd.DataFrame({
    'ID': test_df['ID'],        # must exist in test set
    'price': predictions
})

#Save to CSV
submission.to_csv('../Results/submission.csv', index=False)
```
